```python
# For DataFest 2025 - Savills Commercial Real Estate Data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter
import matplotlib.gridspec as gridspec
from collections import Counter

# Set visualization style
plt.style.use('ggplot')
sns.set_context("talk")

# Define color palettes for sectors and markets
SECTOR_COLORS = {
    "Financial Services and Insurance": "#1f77b4",
    "Financial": "#1f77b4",  # Shortened version
    "Technology": "#ff7f0e",
    "Tech": "#ff7f0e",  # Shortened version
    "Legal Services": "#2ca02c",
    "Legal": "#2ca02c",  # Shortened version
    "Healthcare": "#d62728",
    "Professional Services": "#9467bd",
    "Prof Svcs": "#9467bd",  # Shortened version
    "Media": "#8c564b",
    "Manufacturing": "#e377c2",
    "Manuf": "#e377c2",  # Shortened version
    "Government": "#7f7f7f",
    "Govt": "#7f7f7f",  # Shortened version
    "Education": "#bcbd22",
    "Edu": "#bcbd22",  # Shortened version
    "Retail": "#17becf"
}

# Function to truncate long sector names
def truncate_sector(sector_name):
    if pd.isna(sector_name):
        return np.nan
    
    # Map of full names to shortened versions
    sector_map = {
        "Financial Services and Insurance": "Financial",
        "Technology": "Tech",
        "Legal Services": "Legal",
        "Professional Services": "Prof Svcs",
        "Manufacturing": "Manuf",
        "Government": "Govt",
        "Education": "Edu",
        "Retail and Wholesale": "Retail"
    }
    
    # Return the shortened version if it exists, otherwise return the original
    return sector_map.get(sector_name, sector_name)

# Function to format percentages
def percentage_formatter(x, pos):
    return f'{x:.0f}%'

# Function to format currency
def currency_formatter(x, pos):
    return f'${x:.0f}'

# Load the datasets
print("Loading datasets...")
leases_df = pd.read_csv('Leases.csv')
price_avail_df = pd.read_csv('Price_and_Availability_Data.csv')
occupancy_df = pd.read_csv('Market_Occupancy.csv')

# Basic preprocessing for all datasets
print("Preprocessing data...")

# For leases data
# Convert NA strings to NaN
leases_df = leases_df.replace('NA', np.nan)

# Create a period column for time series analysis
leases_df['period'] = leases_df['year'].astype(str) + '-' + leases_df['quarter']
leases_df['period_num'] = leases_df['year'] + leases_df['quarter'].apply(
    lambda q: {'Q1': 0.0, 'Q2': 0.25, 'Q3': 0.5, 'Q4': 0.75}[q])

# Add shortened sector names
leases_df['sector_short'] = leases_df['internal_industry'].apply(truncate_sector)

# Filter for California markets
# First, identify all California markets
california_markets = [market for market in leases_df['market'].unique() 
                     if market in ['Los Angeles', 'San Francisco', 'San Diego', 'Orange County', 
                                  'Silicon Valley', 'Sacramento', 'East Bay', 'San Jose', 
                                  'South Bay/San Jose', 'Oakland']]

# Filter for California data
ca_leases_df = leases_df[leases_df['market'].isin(california_markets) | 
                        (leases_df['state'] == 'CA')]

# Extract California data from price_avail_df as well
ca_price_df = price_avail_df[price_avail_df['market'].isin(california_markets)]

# Extract California data from occupancy_df
ca_occupancy_df = occupancy_df[occupancy_df['market'].isin(['Los Angeles', 'San Francisco', 'South Bay/San Jose'])]

print(f"Found {len(ca_leases_df)} lease transactions in California")
print(f"California markets: {sorted(ca_leases_df['market'].unique())}")

# Make sure numeric columns are properly formatted
numeric_cols = ['leasedSF', 'RBA', 'available_space', 'availability_proportion', 
                'internal_class_rent', 'overall_rent', 'zip']
for col in numeric_cols:
    if col in ca_leases_df.columns:
        ca_leases_df[col] = pd.to_numeric(ca_leases_df[col], errors='coerce')

# Focus on large leases (>10,000 sq ft) as recommended in the challenge overview
large_leases = ca_leases_df[ca_leases_df['leasedSF'] >= 10000].copy()
print(f"Number of large leases (â‰¥10,000 sq ft) in California: {len(large_leases)}")

# Create a field for pandemic periods
ca_leases_df['pandemic_period'] = 'Pre-pandemic'
ca_leases_df.loc[((ca_leases_df['year'] == 2020) & (ca_leases_df['quarter'].isin(['Q2', 'Q3', 'Q4']))) | 
               (ca_leases_df['year'] > 2020), 'pandemic_period'] = 'Post-pandemic'

# 1. California Sector Distribution Analysis
print("\n=== CALIFORNIA SECTOR DISTRIBUTION ANALYSIS ===")

# Extract the sectors that have significant presence in California
ca_sectors = ca_leases_df['internal_industry'].value_counts()
top_sectors = ca_sectors[ca_sectors >= 10].index.tolist()

print(f"Top sectors in California by transaction count: {top_sectors}")

# Create a visualization of sector distribution using shortened names
plt.figure(figsize=(12, 8))
sector_counts = ca_leases_df['sector_short'].value_counts().head(10)
sector_counts.plot(kind='bar')
plt.title('Top 10 Sectors in California by Transaction Count', fontsize=16)
plt.xlabel('Sector', fontsize=14)
plt.ylabel('Number of Transactions', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 2. Sector Analysis by Square Footage
print("\n=== SECTOR ANALYSIS BY SQUARE FOOTAGE ===")

# Calculate total leased space by sector
sector_sf = ca_leases_df.groupby('internal_industry')['leasedSF'].sum().sort_values(ascending=False)
sector_sf_top = sector_sf.head(10)

# Create a DataFrame with shortened names
sector_sf_df = pd.DataFrame({'leasedSF': sector_sf_top})
sector_sf_df['sector_short'] = [truncate_sector(s) for s in sector_sf_df.index]
sector_sf_short = sector_sf_df.set_index('sector_short')['leasedSF']

plt.figure(figsize=(12, 8))
sector_sf_short.plot(kind='bar')
plt.title('Top 10 Sectors in California by Total Leased Square Footage', fontsize=16)
plt.xlabel('Sector', fontsize=14)
plt.ylabel('Total Square Footage Leased', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 3. Pandemic Impact by Sector
print("\n=== PANDEMIC IMPACT BY SECTOR ===")

# Compare pre and post pandemic leasing activity by sector with shortened names
# First, create a mapping between original sector names and shortened ones
sector_mapping = {}
for sector in top_sectors:
    sector_mapping[sector] = truncate_sector(sector)

# Group by the shortened sector names
ca_leases_df['sector_short'] = ca_leases_df['internal_industry'].apply(truncate_sector)
pandemic_pivot = pd.pivot_table(
    ca_leases_df[ca_leases_df['internal_industry'].isin(top_sectors)],
    values='leasedSF',
    index='sector_short',
    columns='pandemic_period',
    aggfunc='sum'
).fillna(0)

# Calculate the percent change
if 'Pre-pandemic' in pandemic_pivot.columns and 'Post-pandemic' in pandemic_pivot.columns:
    pandemic_pivot['Percent_Change'] = ((pandemic_pivot['Post-pandemic'] / 
                                      pandemic_pivot['Pre-pandemic']) - 1) * 100
    
    # Sort by percent change
    pandemic_pivot = pandemic_pivot.sort_values('Percent_Change', ascending=False)
    
    # Create a visualization
    plt.figure(figsize=(14, 10))
    pandemic_pivot['Percent_Change'].plot(kind='bar')
    plt.title('Pandemic Impact on Leasing Activity by Sector in California', fontsize=16)
    plt.xlabel('Sector', fontsize=14)
    plt.ylabel('Percent Change in Leased Square Footage', fontsize=14)
    plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    
    # Create a more detailed visualization showing pre and post values
    plt.figure(figsize=(14, 10))
    
    x = np.arange(len(pandemic_pivot.index))
    width = 0.35
    
    plt.bar(x - width/2, pandemic_pivot['Pre-pandemic'] / 1000000, width, 
            label='Pre-pandemic', color='skyblue')
    plt.bar(x + width/2, pandemic_pivot['Post-pandemic'] / 1000000, width,
            label='Post-pandemic', color='salmon')
    
    # Add percent change labels
    for i, (idx, row) in enumerate(pandemic_pivot.iterrows()):
        plt.text(i, max(row['Pre-pandemic'], row['Post-pandemic']) / 1000000 + 0.2, 
                 f"{row['Percent_Change']:.1f}%", 
                 ha='center', va='bottom', fontweight='bold')
    
    plt.title('Pre vs Post Pandemic Leasing Activity by Sector in California', fontsize=16)
    plt.xlabel('Sector', fontsize=14)
    plt.ylabel('Total Square Footage Leased (Millions)', fontsize=14)
    plt.xticks(x, pandemic_pivot.index, rotation=45, ha='right')
    plt.legend()
    plt.tight_layout()
    plt.show()

# 4. Geographic Distribution of Sectors within California
print("\n=== GEOGRAPHIC DISTRIBUTION OF SECTORS WITHIN CALIFORNIA ===")

# Analyze sector concentration by market within California using shortened names
top5_sectors_original = sector_sf.head(5).index.tolist()
top5_sectors_short = [truncate_sector(s) for s in top5_sectors_original]

# Create a mapping for the visualization
sector_market_pivot = pd.pivot_table(
    ca_leases_df[ca_leases_df['internal_industry'].isin(top5_sectors_original)],
    values='leasedSF',
    index='market',
    columns='sector_short',
    aggfunc='sum'
).fillna(0)

# Calculate percentage of each sector for each market
sector_market_pct = sector_market_pivot.copy()
for col in sector_market_pct.columns:
    sector_market_pct[col] = sector_market_pct[col] / sector_market_pct[col].sum() * 100

# Create a visualization
plt.figure(figsize=(14, 10))
sector_market_pct.plot(kind='bar', stacked=True)
plt.title('Sector Distribution by California Market', fontsize=16)
plt.xlabel('Market', fontsize=14)
plt.ylabel('Percentage of Sector\'s Total Leased Space', fontsize=14)
plt.legend(title='Sector', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 5. Zip Code Cluster Analysis in California
print("\n=== ZIP CODE CLUSTER ANALYSIS IN CALIFORNIA ===")

# Check if we have sufficient zip code data
has_zip_data = not ca_leases_df['zip'].isna().all()

if has_zip_data:
    # Filter out rows with missing zip codes
    zip_data = ca_leases_df.dropna(subset=['zip'])
    
    # Convert zip to string to ensure proper handling
    zip_data['zip'] = zip_data['zip'].astype(str)
    zip_data['zip'] = zip_data['zip'].str.split('.').str[0]  # Remove decimals if present
    
    # Group by zip code and sector to identify clusters
    zip_sector_counts = zip_data.groupby(['zip', 'internal_industry']).size().reset_index(name='count')
    
    # For each zip code, find the dominant industry
    zip_dominant_sector = zip_sector_counts.loc[zip_sector_counts.groupby('zip')['count'].idxmax()]
    
    # Count transactions by zip code to find major clusters
    zip_transaction_counts = zip_data.groupby('zip').size().reset_index(name='transactions')
    zip_transaction_counts = zip_transaction_counts.sort_values('transactions', ascending=False)
    
    # Look at top 20 zip codes by transaction volume
    top_zips = zip_transaction_counts.head(20)
    
    print("Top 20 California ZIP codes by transaction volume:")
    for idx, row in top_zips.iterrows():
        zip_code = row['zip']
        transactions = row['transactions']
        
        # Get dominant sector for this zip
        dominant = zip_dominant_sector[zip_dominant_sector['zip'] == zip_code]
        if not dominant.empty:
            dominant_sector = dominant['internal_industry'].iloc[0]
            dominant_count = dominant['count'].iloc[0]
            percent = (dominant_count / transactions) * 100
            
            print(f"ZIP {zip_code}: {transactions} transactions, dominated by {dominant_sector} ({percent:.1f}%)")
    
    # Visualize top 10 zip codes and their sector distribution
    top10_zips = top_zips.head(10)['zip'].tolist()
    
    # Create a pivot table of sector distribution for top zips
    zip_sector_pivot = pd.pivot_table(
        zip_data[zip_data['zip'].isin(top10_zips)],
        values='leasedSF',
        index='zip',
        columns='internal_industry',
        aggfunc='sum'
    ).fillna(0)
    
    # Convert to percentages for better visualization
    zip_sector_pct = zip_sector_pivot.copy()
    zip_sector_pct = zip_sector_pct.div(zip_sector_pct.sum(axis=1), axis=0) * 100
    
    # Select top 5 sectors for clarity
    top5_sectors = sector_sf.head(5).index.tolist()
    zip_sector_pct_filtered = zip_sector_pct[top5_sectors].copy() if all(s in zip_sector_pct.columns for s in top5_sectors) else zip_sector_pct
    
    # Create a visualization
    plt.figure(figsize=(14, 10))
    zip_sector_pct_filtered.plot(kind='bar', stacked=True)
    plt.title('Sector Distribution in Top 10 California ZIP Codes by Transaction Volume', fontsize=16)
    plt.xlabel('ZIP Code', fontsize=14)
    plt.ylabel('Percentage of Total Leased Space', fontsize=14)
    plt.legend(title='Sector', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    
    # Create a heatmap to better visualize sector clusters by zip
    # Select top 15 zip codes and top 8 sectors for clarity
    top15_zips = top_zips.head(15)['zip'].tolist()
    top8_sectors = sector_sf.head(8).index.tolist()
    
    # Filter data
    heatmap_data = zip_data[(zip_data['zip'].isin(top15_zips)) & 
                           (zip_data['internal_industry'].isin(top8_sectors))]
    
    # Create a crosstab of transaction counts
    heatmap_pivot = pd.crosstab(
        heatmap_data['zip'], 
        heatmap_data['internal_industry'],
        values=heatmap_data['leasedSF'],
        aggfunc='sum'
    ).fillna(0)
    
    # Normalize by row for better comparison
    heatmap_norm = heatmap_pivot.div(heatmap_pivot.sum(axis=1), axis=0)
    
    # Create a heatmap
    plt.figure(figsize=(16, 10))
    sns.heatmap(heatmap_norm, cmap='YlGnBu', annot=True, fmt='.2f', linewidths=.5)
    plt.title('Sector Concentration Heatmap by ZIP Code in California', fontsize=16)
    plt.ylabel('ZIP Code', fontsize=14)
    plt.xlabel('Sector', fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
else:
    print("Insufficient ZIP code data for cluster analysis")

# 6. Time Series Analysis by Sector in California
print("\n=== TIME SERIES ANALYSIS BY SECTOR IN CALIFORNIA ===")

# Calculate quarterly leasing activity by sector
# Focus on top 5 sectors for clarity
top5_sectors_original = sector_sf.head(5).index.tolist()
top5_sectors_short = [truncate_sector(s) for s in top5_sectors_original]

# Filter data and use shortened names
time_series_data = ca_leases_df[ca_leases_df['internal_industry'].isin(top5_sectors_original)]

# Create a pivot table with quarters and sectors
time_pivot = pd.pivot_table(
    time_series_data,
    values='leasedSF',
    index='period_num',
    columns='sector_short',
    aggfunc='sum'
).fillna(0)

# Create a visualization
plt.figure(figsize=(16, 10))

# Plot each sector
for sector in top5_sectors_short:
    if sector in time_pivot.columns:
        plt.plot(time_pivot.index, time_pivot[sector], 
                 marker='o', linewidth=2, label=sector)

# Add vertical line for pandemic start
plt.axvline(x=2020.25, color='red', linestyle='--', 
            label='Pandemic Start (Q2 2020)', alpha=0.7)

plt.title('Quarterly Leasing Activity by Sector in California', fontsize=16)
plt.xlabel('Year-Quarter', fontsize=14)
plt.ylabel('Total Square Footage Leased', fontsize=14)
plt.legend(title='Sector', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)

# Format the x-axis
plt.xticks([2018, 2018.25, 2018.5, 2018.75, 
            2019, 2019.25, 2019.5, 2019.75, 
            2020, 2020.25, 2020.5, 2020.75,
            2021, 2021.25, 2021.5, 2021.75,
            2022, 2022.25, 2022.5, 2022.75,
            2023, 2023.25, 2023.5, 2023.75,
            2024, 2024.25, 2024.5],
           ['2018-Q1', 'Q2', 'Q3', 'Q4', 
            '2019-Q1', 'Q2', 'Q3', 'Q4',
            '2020-Q1', 'Q2', 'Q3', 'Q4',
            '2021-Q1', 'Q2', 'Q3', 'Q4',
            '2022-Q1', 'Q2', 'Q3', 'Q4',
            '2023-Q1', 'Q2', 'Q3', 'Q4',
            '2024-Q1', 'Q2', 'Q3'], 
           rotation=45, ha='right')

plt.tight_layout()
plt.show()

# 7. Relation between Market Occupancy and Sector Leasing
print("\n=== MARKET OCCUPANCY AND SECTOR LEASING RELATIONSHIP ===")

# Check if we have overlapping data between occupancy and leasing datasets
ca_occupancy_markets = ca_occupancy_df['market'].unique()
overlapping_markets = [m for m in ca_occupancy_markets if m in ca_leases_df['market'].unique()]

if overlapping_markets:
    # Create a visualization showing occupancy and leasing activity for overlapping markets
    plt.figure(figsize=(16, 12))
    
    # Create a subplot for each overlapping market
    fig, axes = plt.subplots(len(overlapping_markets), 1, figsize=(16, 5*len(overlapping_markets)))
    
    if len(overlapping_markets) == 1:
        axes = [axes]  # Wrap in list for consistent indexing
    
    for i, market in enumerate(overlapping_markets):
        ax = axes[i]
        
        # Get occupancy data for this market
        market_occupancy = ca_occupancy_df[ca_occupancy_df['market'] == market]
        
        # Sort by period for proper time series
        market_occupancy['period_num'] = market_occupancy['year'] + market_occupancy['quarter'].apply(
            lambda q: {'Q1': 0.0, 'Q2': 0.25, 'Q3': 0.5, 'Q4': 0.75}[q])
        market_occupancy = market_occupancy.sort_values('period_num')
        
        # Plot occupancy
        ax.plot(market_occupancy['period_num'], market_occupancy['ending_occupancy_proportion']*100, 
                color='blue', marker='o', linewidth=2, label='Occupancy Rate (%)')
        
        # Create a secondary y-axis for leasing activity
        ax2 = ax.twinx()
        
        # Get leasing data for this market
        market_leasing = ca_leases_df[ca_leases_df['market'] == market]
        
        # Calculate total leasing by period
        lease_by_period = market_leasing.groupby('period_num')['leasedSF'].sum().reset_index()
        
        # Plot leasing activity
        ax2.bar(lease_by_period['period_num'], lease_by_period['leasedSF']/1000, 
                alpha=0.3, color='green', label='Leasing Activity (000s SF)')
        
        # Set labels and title
        ax.set_title(f"{market} - Occupancy vs. Leasing Activity", fontsize=14)
        ax.set_xlabel('Year-Quarter', fontsize=12)
        ax.set_ylabel('Occupancy Rate (%)', fontsize=12, color='blue')
        ax2.set_ylabel('Leasing Activity (000s SF)', fontsize=12, color='green')
        
        # Set axis colors
        ax.tick_params(axis='y', labelcolor='blue')
        ax2.tick_params(axis='y', labelcolor='green')
        
        # Format the x-axis
        ax.set_xticks([2020, 2021, 2022, 2023, 2024])
        ax.set_xticklabels(['2020', '2021', '2022', '2023', '2024'])
        
        # Add a legend
        lines1, labels1 = ax.get_legend_handles_labels()
        lines2, labels2 = ax2.get_legend_handles_labels()
        ax.legend(lines1 + lines2, labels1 + labels2, loc='upper left')
        
        # Add grid
        ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# 8. Comprehensive California Real Estate Dashboard
print("\n=== CALIFORNIA REAL ESTATE DASHBOARD ===")

# Create a dashboard with 4 key visualizations
plt.figure(figsize=(20, 16))
gs = gridspec.GridSpec(2, 2, width_ratios=[1, 1], height_ratios=[1, 1])

# Plot 1: California Market Sizes
ax1 = plt.subplot(gs[0, 0])

# Calculate market sizes
ca_market_size = ca_leases_df.groupby('market')['leasedSF'].sum().sort_values(ascending=False)
top_ca_markets = ca_market_size.head(8)

top_ca_markets.plot(kind='barh', ax=ax1, color='skyblue')
ax1.set_title('California Markets by Total Leased Square Footage', fontsize=14)
ax1.set_xlabel('Total Leased Square Footage', fontsize=12)
ax1.set_ylabel('Market', fontsize=12)
ax1.grid(axis='x', alpha=0.3)

# Plot 2: Sector Distribution in California
ax2 = plt.subplot(gs[0, 1])

# Calculate sector distribution with shortened names
ca_sector_dist = ca_leases_df.groupby('sector_short')['leasedSF'].sum().sort_values(ascending=False).head(5)
ca_sector_dist.plot(kind='pie', ax=ax2, autopct='%1.1f%%', startangle=90, 
                    colors=[SECTOR_COLORS.get(s, 'gray') for s in ca_sector_dist.index])
ax2.set_title('Top 5 Sectors in California by Leased Square Footage', fontsize=14)
ax2.set_ylabel('')

# Plot 3: Time Series for Top Markets
ax3 = plt.subplot(gs[1, 0])

# Calculate leasing activity by period for top markets
top_markets_list = top_ca_markets.head(5).index.tolist()
for market in top_markets_list:
    market_data = ca_leases_df[ca_leases_df['market'] == market]
    market_by_period = market_data.groupby('period_num')['leasedSF'].sum()
    ax3.plot(market_by_period.index, market_by_period.values, 
             marker='o', linewidth=2, label=market)

# Add vertical line for pandemic start
ax3.axvline(x=2020.25, color='red', linestyle='--', 
           label='Pandemic Start (Q2 2020)', alpha=0.7)

ax3.set_title('Leasing Activity in Top California Markets Over Time', fontsize=14)
ax3.set_xlabel('Year-Quarter', fontsize=12)
ax3.set_ylabel('Total Square Footage Leased', fontsize=12)
ax3.legend(fontsize=10)
ax3.grid(True, alpha=0.3)

# Format the x-axis
ax3.set_xticks([2018, 2019, 2020, 2021, 2022, 2023, 2024])
ax3.set_xticklabels(['2018', '2019', '2020', '2021', '2022', '2023', '2024'])

# Plot 4: Pandemic Impact on Sectors
ax4 = plt.subplot(gs[1, 1])

if 'Percent_Change' in pandemic_pivot.columns:
    # Use the pandemic impact data calculated earlier
    pandemic_pivot['Percent_Change'].head(8).plot(kind='barh', ax=ax4, color='salmon')
    ax4.axvline(x=0, color='black', linestyle='-', alpha=0.3)
    ax4.set_title('Pandemic Impact on Top Sectors in California', fontsize=14)
    ax4.set_xlabel('Percent Change in Leased Square Footage', fontsize=12)
    ax4.set_ylabel('Sector', fontsize=12)
    ax4.grid(axis='x', alpha=0.3)

plt.suptitle('California Commercial Real Estate Market Analysis', fontsize=20, y=0.98)
plt.tight_layout()
plt.subplots_adjust(top=0.93)
plt.show()

print("Analysis complete. All California sector and market visualizations generated.")

# 9. Tech Sector vs. Financial Services Comparison in California
print("\n=== TECH VS. FINANCIAL SERVICES IN CALIFORNIA ===")

# Check if we have data for both sectors
has_tech = 'Technology' in ca_leases_df['internal_industry'].values
has_financial = 'Financial Services and Insurance' in ca_leases_df['internal_industry'].values

if has_tech and has_financial:
    # Filter data for these two sectors
    tech_financial_df = ca_leases_df[ca_leases_df['internal_industry'].isin(['Technology', 'Financial Services and Insurance'])]
    
    # Add shortened names
    tech_financial_df['sector_short'] = tech_financial_df['internal_industry'].apply(truncate_sector)
    
    # Compare average lease size
    avg_lease_size = tech_financial_df.groupby('sector_short')['leasedSF'].mean()
    print(f"Average lease size by sector:\n{avg_lease_size}")
    
    # Compare building class preference
    class_preference = pd.crosstab(
        tech_financial_df['sector_short'], 
        tech_financial_df['internal_class'],
        values=tech_financial_df['leasedSF'],
        aggfunc='sum',
        normalize='index'
    ) * 100
    
    print(f"Building class preference by sector (percentage of total leased SF):\n{class_preference}")
    
    # Compare geographic distribution
    geo_distribution = pd.crosstab(
        tech_financial_df['sector_short'], 
        tech_financial_df['market'],
        values=tech_financial_df['leasedSF'],
        aggfunc='sum',
        normalize='index'
    ) * 100
    
    print(f"Geographic distribution by sector (percentage of total leased SF):")
    for sector in geo_distribution.index:
        top_markets = geo_distribution.loc[sector].sort_values(ascending=False).head(3)
        print(f"{sector} - Top 3 markets: {', '.join([f'{m}: {v:.1f}%' for m, v in top_markets.items()])}")
    
    # Compare time trends
    time_trends = pd.pivot_table(
        tech_financial_df,
        values='leasedSF',
        index='period_num',
        columns='sector_short',
        aggfunc='sum'
    ).fillna(0)
    
    # Create visualizations
    fig, axes = plt.subplots(2, 2, figsize=(18, 14))
    
    # Plot 1: Average Lease Size
    avg_lease_size.plot(kind='bar', ax=axes[0, 0], color=['#ff7f0e', '#1f77b4'])
    axes[0, 0].set_title('Average Lease Size Comparison', fontsize=14)
    axes[0, 0].set_ylabel('Square Footage', fontsize=12)
    axes[0, 0].set_xlabel('', fontsize=12)
    axes[0, 0].grid(axis='y', alpha=0.3)
    
    # Plot 2: Building Class Preference
    class_preference.plot(kind='bar', ax=axes[0, 1], stacked=True)
    axes[0, 1].set_title('Building Class Preference', fontsize=14)
    axes[0, 1].set_ylabel('Percentage of Leased Space', fontsize=12)
    axes[0, 1].set_xlabel('', fontsize=12)
    axes[0, 1].grid(axis='y', alpha=0.3)
    
    # Plot 3: Geographic Distribution
    # Select top 5 markets for clarity
    top_markets_all = geo_distribution.sum().sort_values(ascending=False).head(5).index
    geo_dist_filtered = geo_distribution[top_markets_all]
    
    geo_dist_filtered.plot(kind='bar', ax=axes[1, 0])
    axes[1, 0].set_title('Market Distribution Comparison', fontsize=14)
    axes[1, 0].set_ylabel('Percentage of Leased Space', fontsize=12)
    axes[1, 0].set_xlabel('', fontsize=12)
    axes[1, 0].grid(axis='y', alpha=0.3)
    
    # Plot 4: Time Trends
    time_trends.plot(ax=axes[1, 1], marker='o')
    axes[1, 1].axvline(x=2020.25, color='red', linestyle='--', label='Pandemic Start', alpha=0.7)
    axes[1, 1].set_title('Leasing Activity Over Time', fontsize=14)
    axes[1, 1].set_ylabel('Square Footage Leased', fontsize=12)
    axes[1, 1].set_xlabel('Year-Quarter', fontsize=12)
    axes[1, 1].grid(True, alpha=0.3)
    
    # Format x-axis
    axes[1, 1].set_xticks([2018, 2019, 2020, 2021, 2022, 2023, 2024])
    axes[1, 1].set_xticklabels(['2018', '2019', '2020', '2021', '2022', '2023', '2024'])
    
    plt.suptitle('Technology vs. Financial Services in California', fontsize=18, y=0.98)
    plt.tight_layout()
    plt.subplots_adjust(top=0.93)
    plt.show()
    
    # Additional Analysis: Transaction Types by Sector
    transaction_pivot = pd.crosstab(
        tech_financial_df['internal_industry'],
        tech_financial_df['transaction_type'],
        values=tech_financial_df['leasedSF'],
        aggfunc='sum',
        normalize='index'
    ) * 100
    
    plt.figure(figsize=(14, 8))
    transaction_pivot.plot(kind='bar', stacked=True)
    plt.title('Transaction Types by Sector', fontsize=16)
    plt.xlabel('Sector', fontsize=14)
    plt.ylabel('Percentage of Leased Space', fontsize=14)
    plt.grid(axis='y', alpha=0.3)
    plt.legend(title='Transaction Type', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
else:
    print("Insufficient data for Tech vs Financial Services comparison")

# 10. ZIP Code Analysis for Tech Companies
print("\n=== ZIP CODE ANALYSIS FOR TECH COMPANIES ===")

if has_zip_data and has_tech:
    # Filter for tech companies
    tech_df = ca_leases_df[ca_leases_df['internal_industry'] == 'Technology'].copy()
    tech_df = tech_df.dropna(subset=['zip'])
    
    # Convert zip to string and clean
    tech_df['zip'] = tech_df['zip'].astype(str)
    tech_df['zip'] = tech_df['zip'].str.split('.').str[0]  # Remove decimals if present
    
    # Count tech transactions by zip
    tech_zip_counts = tech_df.groupby('zip').size().reset_index(name='transactions')
    tech_zip_counts = tech_zip_counts.sort_values('transactions', ascending=False)
    
    # Find total square footage by zip
    tech_zip_sf = tech_df.groupby('zip')['leasedSF'].sum().reset_index()
    tech_zip_sf = tech_zip_sf.sort_values('leasedSF', ascending=False)
    
    # Get top tech zip codes
    top_tech_zips = tech_zip_sf.head(10)
    
    print("Top 10 ZIP codes for Technology sector in California:")
    for idx, row in top_tech_zips.iterrows():
        print(f"ZIP {row['zip']}: {row['leasedSF']:,.0f} SF leased")
    
    # Compare with Financial Services
    if has_financial:
        # Filter for financial companies
        financial_df = ca_leases_df[ca_leases_df['internal_industry'] == 'Financial Services and Insurance'].copy()
        financial_df = financial_df.dropna(subset=['zip'])
        
        # Convert zip to string and clean
        financial_df['zip'] = financial_df['zip'].astype(str)
        financial_df['zip'] = financial_df['zip'].str.split('.').str[0]
        
        # Find total square footage by zip
        financial_zip_sf = financial_df.groupby('zip')['leasedSF'].sum().reset_index()
        financial_zip_sf = financial_zip_sf.sort_values('leasedSF', ascending=False)
        
        # Get top financial zip codes
        top_financial_zips = financial_zip_sf.head(10)
        
        print("\nTop 10 ZIP codes for Financial Services sector in California:")
        for idx, row in top_financial_zips.iterrows():
            print(f"ZIP {row['zip']}: {row['leasedSF']:,.0f} SF leased")
        
        # Create visualization
        top_tech_zips_list = top_tech_zips['zip'].tolist()
        top_financial_zips_list = top_financial_zips['zip'].tolist()
        all_top_zips = list(set(top_tech_zips_list + top_financial_zips_list))
        
        # Prepare data for sector comparison by zip
        tech_zip_dict = tech_zip_sf.set_index('zip')['leasedSF'].to_dict()
        financial_zip_dict = financial_zip_sf.set_index('zip')['leasedSF'].to_dict()
        
        # Create comparison dataframe
        zip_comparison = []
        for zip_code in all_top_zips:
            zip_comparison.append({
                'zip': zip_code,
                'Technology': tech_zip_dict.get(zip_code, 0),
                'Financial Services': financial_zip_dict.get(zip_code, 0)
            })
        
        zip_comparison_df = pd.DataFrame(zip_comparison)
        
        # Calculate percentage - which sector dominates each zip
        zip_comparison_df['Total'] = zip_comparison_df['Technology'] + zip_comparison_df['Financial Services']
        zip_comparison_df['Tech_Pct'] = zip_comparison_df['Technology'] / zip_comparison_df['Total'] * 100
        zip_comparison_df['Financial_Pct'] = zip_comparison_df['Financial Services'] / zip_comparison_df['Total'] * 100
        
        # Sort by total leased space
        zip_comparison_df = zip_comparison_df.sort_values('Total', ascending=False).head(15)
        
        # Create a stacked bar chart
        plt.figure(figsize=(14, 10))
        
        # Create stacked bars
        zip_comparison_df_plot = zip_comparison_df.sort_values('Tech_Pct')
        plt.barh(zip_comparison_df_plot['zip'], zip_comparison_df_plot['Tech_Pct'], 
                color='#ff7f0e', label='Technology')
        plt.barh(zip_comparison_df_plot['zip'], zip_comparison_df_plot['Financial_Pct'], 
                left=zip_comparison_df_plot['Tech_Pct'], color='#1f77b4', 
                label='Financial Services')
        
        plt.title('Technology vs. Financial Services by ZIP Code', fontsize=16)
        plt.xlabel('Percentage of Total Leased Space', fontsize=14)
        plt.ylabel('ZIP Code', fontsize=14)
        plt.legend(fontsize=12)
        plt.grid(axis='x', alpha=0.3)
        plt.tight_layout()
        plt.show()
        
        # Look for overlap in top zip codes
        common_zips = set(top_tech_zips_list).intersection(set(top_financial_zips_list))
        if common_zips:
            print(f"\nZIP codes with strong presence of both Technology and Financial Services: {common_zips}")
        else:
            print("\nNo common ZIP codes found in top 10 for both sectors")
else:
    print("Insufficient data for Technology sector ZIP code analysis")

# 11. Time Series Analysis for California vs. Rest of US
print("\n=== CALIFORNIA VS. REST OF US COMPARISON ===")

# Create a Rest of US dataset
non_ca_leases_df = leases_df[~leases_df['market'].isin(california_markets) & 
                           (leases_df['state'] != 'CA')]

# Calculate total leasing by period for California and Rest of US
ca_time_series = ca_leases_df.groupby('period_num')['leasedSF'].sum().reset_index()
non_ca_time_series = non_ca_leases_df.groupby('period_num')['leasedSF'].sum().reset_index()

# Calculate the ratio of California to total US leasing
ca_time_series['ca_ratio'] = ca_time_series['leasedSF'] / (ca_time_series['leasedSF'] + non_ca_time_series['leasedSF'])

# Create visualization
plt.figure(figsize=(16, 10))

# Plot California and Rest of US leasing
ax1 = plt.subplot(211)
ax1.plot(ca_time_series['period_num'], ca_time_series['leasedSF'] / 1000000, 
        marker='o', linewidth=2, label='California', color='gold')
ax1.plot(non_ca_time_series['period_num'], non_ca_time_series['leasedSF'] / 1000000, 
        marker='o', linewidth=2, label='Rest of US', color='royalblue')

# Add vertical line for pandemic start
ax1.axvline(x=2020.25, color='red', linestyle='--', 
            label='Pandemic Start (Q2 2020)', alpha=0.7)

ax1.set_title('Leasing Activity: California vs. Rest of US', fontsize=16)
ax1.set_ylabel('Total Square Footage Leased (Millions)', fontsize=14)
ax1.legend(fontsize=12)
ax1.grid(True, alpha=0.3)

# Format the x-axis
ax1.set_xticks([2018, 2019, 2020, 2021, 2022, 2023, 2024])
ax1.set_xticklabels(['2018', '2019', '2020', '2021', '2022', '2023', '2024'])

# Plot California's percentage of total US leasing
ax2 = plt.subplot(212)
ax2.plot(ca_time_series['period_num'], ca_time_series['ca_ratio'] * 100, 
        marker='o', linewidth=2, color='forestgreen')

# Add vertical line for pandemic start
ax2.axvline(x=2020.25, color='red', linestyle='--', alpha=0.7)

ax2.set_title('California\'s Share of Total US Leasing Activity', fontsize=16)
ax2.set_xlabel('Year', fontsize=14)
ax2.set_ylabel('Percentage of US Total', fontsize=14)
ax2.grid(True, alpha=0.3)

# Format the x-axis
ax2.set_xticks([2018, 2019, 2020, 2021, 2022, 2023, 2024])
ax2.set_xticklabels(['2018', '2019', '2020', '2021', '2022', '2023', '2024'])

plt.tight_layout()
plt.show()

print("California market sector analysis complete.")
```
